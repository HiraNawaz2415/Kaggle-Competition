## 3. Cleaning of data

```python
## Remove id columns because it has no impact
train.drop("id", axis=1, inplace=True)
test.drop("id", axis=1, inplace=True)
```

```python
print("After dropping id column")
train.head()
```

```python
# -----------------------------
# CLEANING LAP TIME
# -----------------------------

# Limit (clip) LapTime values between 60 and 200 seconds in TRAIN data
# → Removes unrealistic very small or very large lap times (outliers)
train["LapTime (s)"] = train["LapTime (s)"].clip(60, 200)

# Apply same clipping to TEST data (IMPORTANT: train/test must be consistent)
test["LapTime (s)"] = test["LapTime (s)"].clip(60, 200)
```

```python
# -----------------------------
# CLEANING LAP TIME DELTA
# -----------------------------

# Clip LapTime_Delta between -100 and 100 in TRAIN data
# → Removes extreme spikes (e.g., -2000 or +2000) which are unrealistic
# → Keeps model stable and prevents it from learning noise
train["LapTime_Delta"] = train["LapTime_Delta"].clip(-100, 100)

# Apply same transformation to TEST data
test["LapTime_Delta"] = test["LapTime_Delta"].clip(-100, 100)
```

```python
# -----------------------------
# CLEANING DEGRADATION
# -----------------------------

# Clip Cumulative_Degradation between -200 and 200 in TRAIN data
# → Removes abnormal degradation values
# → Ensures model focuses on realistic tire wear patterns
train["Cumulative_Degradation"] = train["Cumulative_Degradation"].clip(-200, 200)

# Apply same clipping to TEST data
test["Cumulative_Degradation"] = test["Cumulative_Degradation"].clip(-200, 200)
```

## 4. Feature Engineering

#### FEATURE ENGINEERING SUMMARY
| Feature Name                    | Formula                                  | Type           | Purpose                           | Intuition (Why it helps)                          |                                      |                                               |
| ------------------------------- | ---------------------------------------- | -------------- | --------------------------------- | ------------------------------------------------- | ------------------------------------ | --------------------------------------------- |
| **TyreEfficiency**              | TyreLife / (LapNumber + 1)               | Ratio          | Measures tire usage efficiency    | Lower value = worn tires → higher pit probability |                                      |                                               |
| **DegradationPressure**         | Cumulative_Degradation / (TyreLife + 1)  | Ratio          | Measures tire stress level        | High pressure = tire failure risk → pit likely    |                                      |                                               |
| **RaceStage**                   | pd.cut(LapNumber, bins=3)                | Categorical    | Splits race into early/mid/late   | Pit strategy changes across race phases           |                                      |                                               |
| **LapStress**                   | LapTime +                                | LapTime_Delta  |                                   | Combined                                          | Measures lap performance instability | High stress = tire degradation → pit decision |
| **PositionPressure**            | Position × RaceProgress                  | Interaction    | Combines position and race timing | Late race + poor position = strategic pit         |                                      |                                               |
| **Aggression**                  | Position_Change × LapNumber              | Interaction    | Captures driving volatility       | Frequent changes = strategy adjustment needed     |                                      |                                               |
| **PitSignal**                   | PitStop × LapNumber                      | Interaction    | Encodes past pit behavior         | Previous pit history influences next decision     |                                      |                                               |
| **TyreAgeRatio**       | TyreLife / (Stint + 1)                   | Ratio          | Measures tyre usage per stint     | Higher value = long tire usage → pit risk         |                                      |                                               |
| **NormalizedPosition** | Position / 20                            | Scaled Feature | Standardizes race position        | Makes position comparable across races            |                                      |                                               |
| **DegradationTrend**   | Cumulative_Degradation / (LapNumber + 1) | Trend Feature  | Measures degradation over time    | Rising trend = tire performance dropping          |                                      |                                               |



These engineered features are inspired by real Formula 1 race strategy dynamics.

Pit stop decisions are influenced by:
- Tire wear and degradation
- Race stage (early vs late strategy shifts)
- Driver behavior and position pressure
- Historical pit strategy patterns

By transforming raw race data into structured signals.The model can better understand hidden racing strategies and improve prediction accuracy for PitNextLap.

```python
# -----------------------------
# 1. TYRE EFFICIENCY
# -----------------------------

# Measures how efficiently the tire is being used
# = how much tyre life is left relative to laps completed
# Low value → tire is worn out → pit likely
train["TyreEfficiency"] = train["TyreLife"] / (train["LapNumber"] + 1)
test["TyreEfficiency"] = test["TyreLife"] / (test["LapNumber"] + 1)
```

```python
# -----------------------------
# 2. DEGRADATION PRESSURE
# -----------------------------

# Measures how much degradation pressure is on the tire
# Combines tire wear (degradation) with current tyre life
# High value → tire under stress → pit likely
train["DegradationPressure"] = train["Cumulative_Degradation"] / (train["TyreLife"] + 1)
test["DegradationPressure"] = test["Cumulative_Degradation"] / (test["TyreLife"] + 1)
```

```python
# -----------------------------
# 3. RACE STAGE (EARLY / MID / LATE)
# -----------------------------

# Divides race into 3 stages:
# 0 = early race
# 1 = mid race
# 2 = late race
# Pit strategy depends heavily on race stage
train["RaceStage"] = pd.cut(train["LapNumber"], bins=3, labels=[0,1,2])
test["RaceStage"] = pd.cut(test["LapNumber"], bins=3, labels=[0,1,2])
```

```python
# -----------------------------
# 4. LAP STRESS
# -----------------------------

# Measures how stressful the lap is on the car/tire
# Combines:
# - actual lap time
# - change in lap time (delta)
# High value → performance drop → pit likely
train["LapStress"] = train["LapTime (s)"] + abs(train["LapTime_Delta"])
test["LapStress"] = test["LapTime (s)"] + abs(test["LapTime_Delta"])
```

```python
# -----------------------------
# 5. POSITION PRESSURE
# -----------------------------

# Combines race position with race progress
# Example:
# - Late race + bad position → high pressure → pit strategy change
train["PositionPressure"] = train["Position"] * train["RaceProgress"]
test["PositionPressure"] = test["Position"] * test["RaceProgress"]
```

```python
# -----------------------------
# 6. AGGRESSION FEATURE
# -----------------------------

# Measures how aggressively driver is changing position over time
# Large position changes → strategic decisions → pit more likely
train["Aggression"] = train["Position_Change"] * train["LapNumber"]
test["Aggression"] = test["Position_Change"] * test["LapNumber"]
```

```python
# -----------------------------
# 7. PIT SIGNAL (STRATEGY MEMORY)
# -----------------------------

# Captures influence of previous pit stops
# If driver already pitted recently → affects next pit decision
train["PitSignal"] = train["PitStop"] * train["LapNumber"]
test["PitSignal"] = test["PitStop"] * test["LapNumber"]
```

```python
# -----------------------------
# 8. TYRE AGE RATIO
# -----------------------------
# How old the tyre is compared to stint usage
# High value → tire heavily used → pit likely
train["TyreAgeRatio"] = train["TyreLife"] / (train["Stint"] + 1)
test["TyreAgeRatio"] = test["TyreLife"] / (test["Stint"] + 1)
```

```python
# -----------------------------
# 9. NORMALIZED POSITION
# -----------------------------
# Scales position into 0–1 range
# Lower position = better rank
train["NormalizedPosition"] = train["Position"] / 20
test["NormalizedPosition"] = test["Position"] / 20
```

```python
# -----------------------------
# 10. DEGRADATION TREND
# -----------------------------
# Shows how degradation is increasing over laps
# High trend → tire worsening fast → pit likely
train["DegradationTrend"] = train["Cumulative_Degradation"] / (train["LapNumber"] + 1)
test["DegradationTrend"] = test["Cumulative_Degradation"] / (test["LapNumber"] + 1)
```

```python
# -----------------------------
# Tyre Efficiency cleaning
# -----------------------------

# Clip TyreEfficiency between 0 and 2 in TRAIN data
# Why? Efficiency should be in a realistic range; values too high are abnormal
train["TyreEfficiency"] = train["TyreEfficiency"].clip(0, 2)

# Apply same rule to TEST data
test["TyreEfficiency"] = test["TyreEfficiency"].clip(0, 2)
```
